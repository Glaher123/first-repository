```python
new_string = "Hello World "
print(new_string)

print(new_string[3])
```

```python
#little bit slicing 
print(new_string[3:])
```

```python
#Reversing the string 
print(new_string[::-1])
```

```python
#Findind the len and couting the occurances 
print(len(new_string))
print(new_string.count("l"))
```

```python
#lowercase

my_string = new_string.lower()
print(my_string)
```

```python
#titile that is to say to turn digets turn bigger after space 
modified_str = my_string.title()
print(modified_str)
```

```python
#code for searching index 
index = new_string.find("w")
print(index)
```

```python
#replacing a substring 

replaced = new_string.replace("World","Friend") #World is to change with the friend
print(replaced)
```

```python
#splitting a string 
splitted = new_string.split() #split by default 
print(splitted)

#for underscore 
modifiedsplit = new_string.split("d") #it will split by the value that you put in the string 
print(modifiedsplit)
```
